#### Create a new column in this data which is the percent change between the open and closing prices of each ticker on each day

```python
stocks["intraday"] = ((stocks["price_close"]-stocks["price_open"]) / stocks["price_open"])
```

#### Report the 5 largest intraday *drops* in this dataset

```python
stocks.sort_values(by = "intraday").head()
```

#### Remove this new intraday column, along with the employees column, from the data; print the result to verify they have indeed been dropped

```python
stocks_new = stocks.drop(["employees", "intraday"], axis=1)
```

```python
stocks_new.head()
```

#### Find the standard deviation of each ticker's closing stock price across this time horizon; sort them highest to lowest

```python
standard_deviation = stocks[["ticker", "price_close"]].groupby("ticker").std()
standard_deviation.sort_values("price_close", ascending = False)
```

#### Create daily, sector-level averages of the price columns

```python
groupdata = stocks_new.groupby(["date", "sector"]).mean(numeric_only=True)
groupdata
```

#### Using these sector averages, which sector, on average, experiences the largest percent change between daily high and low prices?

Calculate this as $\frac{high - low}{low}$

```python
groupdata['high_low'] = (groupdata['price_high']-groupdata['price_low'])/groupdata['price_low']
groupdata2 = groupdata.groupby("sector").mean(numeric_only=True)
groupdata2 = groupdata2.drop(["price_open", "price_high","price_low","price_close","volume"], axis=1)
final = groupdata2.sort_values("high_low", ascending = False)
final.reset_index().head(1)
```
